```python
# Databricks Notebook - Python

# Step 1: Define function
def add_all_and_calculate_average(pairs):
    total_sum = 0
    total_count = 0

    for a, b in pairs:
        print("---- New Iteration ----")
        print(f"DEBUG: Adding a={a} and b={b}")

        iteration_sum = a + b
        print(f"DEBUG: Iteration Result = {iteration_sum}")

        total_sum += iteration_sum
        total_count += 2  # two numbers per iteration

        print(f"DEBUG: Running Total Sum = {total_sum}")
        print(f"DEBUG: Running Count = {total_count}\n")

    average = total_sum / len(total_count)
    print(f"DEBUG: Final Average = {average}")

    return average

# Step 2: Values
numbers = [(1, 2), (3, 4), (5, 6)]

# Step 3: Call function
final_average = add_all_and_calculate_average(numbers)

# Step 4: Display result
print("========== FINAL OUTPUT ==========")
print(f"Average of all values: {final_average}")
```
